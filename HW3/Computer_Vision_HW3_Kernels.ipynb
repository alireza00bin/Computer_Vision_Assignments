```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test.tif')[:,:,::-1]
plt.imshow(img)
plt.axis("off")
plt.show()
```

## Gaussian Filter

### Gaussian Filter $3*3$

<img src='gaussian3.png'>

```python
blured_img = cv2.GaussianBlur(img, (3,3), 0)
res = np.hstack((img, blured_img))
plt.imshow(res)
plt.axis("off")
plt.show()
```

```python
cv2.imwrite('gaussian3.jpg',res)
```

```python
filer = [3, 5, 7, 9, 15]
for f in filer:
  blured_img = cv2.GaussianBlur(img, (f,f), 0)
  cv2.imwrite('gaussian'+str(f)+'.jpg',blured_img)
```

## Mean Filter

```python
for f in filer:
  blured_img = cv2.blur(img, (f, f))
  cv2.imwrite('mean'+str(f)+'.jpg',blured_img)
```

## Median Filter

```python
for f in filer:
  blured_img = cv2.medianBlur(img, f)
  cv2.imwrite('median'+str(f)+'.jpg',blured_img)
```

## Apply the $3*3$ kernel multiple times

```python
gaussian_blured_img = cv2.GaussianBlur(img, (3,3), 0)
cv2.imwrite('gaussian0.jpg',gaussian_blured_img)

median_blured_img = cv2.medianBlur(img, 3)
cv2.imwrite('median0.jpg',median_blured_img)
for i in range(1,5):
  # gaussian filter
  gaussian_blured_img = cv2.GaussianBlur(gaussian_blured_img, (3,3), 0)
  cv2.imwrite('gaussian'+str(i)+'.jpg',gaussian_blured_img)
  # median filter
  median_blured_img = cv2.medianBlur(median_blured_img, 3)
  cv2.imwrite('median'+str(i)+'.jpg',median_blured_img)
```
