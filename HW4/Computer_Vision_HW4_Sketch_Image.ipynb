```python
import cv2 as cv
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

### Loading and Plotting Original Image

```python
img = cv.imread("image.jpg")[:,:,::-1]
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

### Converting Image to GrayScale

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
cv.imwrite('GrayScaleImage.jpg', img_gray)
```

### Inverting the Image

```python
img_invert = cv.bitwise_not(img_gray)
plt.imshow(img_invert, cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
cv.imwrite('InvertedImage.jpg', img_invert)
```

### Smoothing image

```python
img_smoothing = cv.GaussianBlur(img_invert, (21, 21), sigmaX=0, sigmaY=0)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
cv.imwrite('SmoothenImage.jpg', img_smoothing)
```

### Converting your image to pencil sketches

```python
final = cv.divide(img_gray, 255 - img_smoothing, scale=255)
plt.imshow(final, cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
cv.imwrite('SketchImage.jpg', final)
```

```python
f = plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_invert, cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smoothing, cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.savefig('Conclusion.jpg')
plt.show()
```
